# 王勖成《有限单元法》部分练习题讲解

## 第2章 弹性力学问题有限单元法的一般原理和表达格式

### 练习题 2.6

    验证3结点三角形单元发生刚体位移时，单元中不产生应力，即赋予结点在单元做平动和转动时相应的结点位移，单元中的应力为零。

##  关键问题

- 刚体位移的分类 —— 平动与转动
- 刚体位移的表示形式

## 1  弹性力学理论

### 1.1  平动刚体位移

$$
\left\{
\begin{array}{rcl}
u_t(x,y) &=& u_0 \\
v_t(x,y) &=& v_0
\end{array}
\right.
\tag{1}
$$

### 1.2  转动刚体位移

绕参考点$(x_0,y_0)$，转角为$\theta$。
$$
\left\{
\begin{array}{rcr}
u_r(x,y) & = &-(y-y_0)\theta \\
v_r(x,y) & = & (x-x_0)\theta
\end{array}
\right.
\tag{2}
$$

### 1.3  一般刚体位移

$$
\left\{
\begin{array}{rcl}
u(x,y) &=& u_0 - (y-y_0)\theta \\
v(x,y) &=& v_0 + (x-x_0)\theta
\end{array}
\right.
\tag{3}
$$

### 1.4  几何方程求应变

对于式$(3)$给出的一般刚体位移函数，由<font color="#660000">几何方程</font>，可求得相应的应变为
$$
\left\{
\begin{array}{l}
\varepsilon_x = \dfrac{\partial u}{\partial x} = 0 \\
\varepsilon_y = \dfrac{\partial v}{\partial y} = 0 \\
\gamma_{xy} = \dfrac{\partial u}{\partial y} + \dfrac{\partial v}{\partial x} = 0
\end{array}
\right.
\tag{4}
$$

## 2  有限元方法

### 2.1  3结点三角形单元

假定平面3结点三角形单元的结点坐标分别为$(x_i,y_i)$、$(x_j,y_j)$和$(x_m,y_m)$。

### 2.2  3结点三角形单元的刚体平动位移

与刚体平动位移$(u_0,v_0)$对应的单元结点位移向量为

$$
\boldsymbol{a}^t
=
\left\{
\begin{array}{c}
u_i^t\\
v_i^t\\
u_j^t\\
v_j^t\\
u_m^t\\
v_m^t
\end{array}
\right\}
=
\left\{
\begin{array}{c}
u_0\\
v_0\\
u_0\\
v_0\\
u_0\\
v_0
\end{array}
\right\}
\tag{5}
$$

### 2.3  3结点三角形单元的刚体转动位移

假定单元绕参考点$(x_0,y_0)$，做转角为$\theta$的刚体转动，对应的单元结点位移向量为
$$
\boldsymbol{a}^r
=
\left\{
\begin{array}{c}
u_i^r\\
v_i^r\\
u_j^r\\
v_j^r\\
u_m^r\\
v_m^r
\end{array}
\right\}
=
\theta\left\{
\begin{array}{c}
-y_i+y_0\\
x_i-x_0\\
-y_j+y_0\\
x_j-x_0\\
-y_m+y_0\\
x_m-x_0
\end{array}
\right\}
\tag{6}
$$

### 2.4  3结点三角形单元的一般刚体位移

$$
\boldsymbol{a} = \boldsymbol{a}^t + \boldsymbol{a}^r\tag{7}
$$

或

$$
\left\{
\begin{array}{c}
u_i\\
v_i\\
u_j\\
v_j\\
u_m\\
v_m
\end{array}
\right\}
=
\left\{
\begin{array}{c}
u_0\\
v_0\\
u_0\\
v_0\\
u_0\\
v_0
\end{array}
\right\}
+
\theta\left\{
\begin{array}{c}
-y_i+y_0\\
x_i-x_0\\
-y_j+y_0\\
x_j-x_0\\
-y_m+y_0\\
x_m-x_0
\end{array}
\right\}
\tag{8}
$$

### 2.5  3结点三角形单元的应变矩阵

$$
\boldsymbol{B}
=\dfrac{1}{2A}
\left[
\begin{array}{cc|cc|cc}
  b_i  & 0   & b_j  & 0   & b_m  & 0  \\
  0   & c_i  & 0   & c_j  & 0   & c_m \\
  c_i  & b_i  & c_j  & b_j  & c_m  & b_m
\end{array}
\right]
\tag{9}
$$

其中

$$
\left\{
\begin{array}{rcl}
b_i & = & -\left|\begin{array}{cc}
       1 & y_j\\
       1 & y_m
     \end{array}\right|
   =  y_j-y_m               \\
b_j & = & -\left|
     \begin{array}{cc}
       1 & y_m\\
       1 & y_i
     \end{array}\right|
   =  y_m-y_i               \\
b_m & = & -\left|\begin{array}{cc}
       1 & y_i\\
       1 & y_j
     \end{array}\right|
   =  y_i-y_j
\end{array}
\right.
\tag{10}
$$


$$
\left\{
\begin{array}{rcl}
c_i & = & \left|\begin{array}{cc}
       1 & x_j\\
       1 & x_m
     \end{array}\right|
   = -x_j+x_m               \\
c_j & = & \left|\begin{array}{cc}
       1 & x_m\\
       1 & x_i
    \end{array}\right|
   = -x_m+x_i               \\
c_m & = & \left|\begin{array}{cc}
       1 & x_i\\
       1 & x_j
     \end{array}\right|
   = -x_i+x_j
\end{array}
\right.
\tag{11}
$$

### 2.6  单元应变 —— 采用应变矩阵和单元结点位移向量表示

可由应变矩阵$\boldsymbol{B}$和单元结点位移向量$\boldsymbol{a}$求单元应变式

$$
\boldsymbol{\varepsilon} = \boldsymbol{B}\boldsymbol{a}\tag{12}
$$

或

$$
\left\{\begin{array}{c}
      \varepsilon_x\\
      \varepsilon_y\\
      \gamma_{xy}
\end{array}\right\}
=\dfrac{1}{2A}
\left[\begin{array}{cccccc}
      b_i & 0   & b_j & 0   & b_m & 0   \\
      0   & c_i & 0   & c_j & 0   & c_m \\
      c_i & b_i & c_j & b_j & c_m & b_m
\end{array}\right]
\left\{\begin{array}{c}
u_i\\
v_i\\
u_j\\
v_j\\
u_m\\
v_m
\end{array}\right\}
\tag{13}
$$

由式$(10)$和式$(11)$，可知
 
$$
b_i + b_j + b_m
=
-\left|
\begin{array}{ccc}
   1 & 1 & y_i\\
   1 & 1 & y_j\\
   1 & 1 & y_m
\end{array}
\right|
=0
\tag{14}
$$

$$
c_i + c_j + c_m
=
\left|
\begin{array}{ccc}
   1 & x_i & 1\\
   1 & x_j & 1\\
   1 & x_m & 1
\end{array}
\right|
=0
\tag{15}
$$

$$
y_ib_i + y_jb_j +y_mb_m
=
\left|
\begin{array}{ccc}
   1 & y_i & y_i\\
   1 & y_j & y_j\\
   1 & y_m & y_m
\end{array}
\right|
=0
\tag{16}
$$

$$
x_ic_i + x_jc_j +x_mc_m
=
\left|
\begin{array}{ccc}
   1 & x_i & x_i\\
   1 & x_j & x_j\\
   1 & x_m & x_m
\end{array}
\right|
=0
\tag{17}
$$

以及


$$
x_ib_i + x_jb_j +x_mb_m
=
\left|
\begin{array}{ccc}
   1 & x_i & y_i\\
   1 & x_j & y_j\\
   1 & x_m & y_m
\end{array}
\right|
=2A
\tag{18}
$$

$$
y_ic_i + y_jc_j +y_mc_m
=
\left|
\begin{array}{ccc}
   1 & x_i & y_i\\
   1 & x_j & y_j\\
   1 & x_m & y_m
\end{array}
\right|
=2A
\tag{19}
$$

### 3    综合结论

将式$(8)$及式$(14\sim 19)$代入式$(13)$，展开计算可得单元内应变

$$
\left\{\begin{array}{c}
\varepsilon_x\\
\varepsilon_y\\
\gamma_{xy}
\end{array}\right\}
=
\left\{\begin{array}{c}
0\\
0\\
0
\end{array}\right\}
\tag{20}
$$

进而，由本构关系可得单元内的应力也为零。

综上所述，3结点三角形单元发生刚体位移时，单元中不会产生应力。
