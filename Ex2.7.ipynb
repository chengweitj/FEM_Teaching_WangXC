# 王勖成《有限单元法》部分练习题讲解

## 第2章 弹性力学问题有限单元法的一般原理和表达格式

### 练习题 2.7

- 二维单元在$x$, $y$坐标平面内，平移到不同位置，单元刚度矩阵相同吗？
- 在平面内旋转时怎样？
- 单元旋转180度后怎样？
- 单元做上述变化时，应力矩阵$\boldsymbol{S}$如何变化？

##  说明

许多同学可能是按网上流传的答案，针对3结点三角形单元来做，结论或许对，但结论不严谨。

## 物理性质

单元发生任何移动、转动，变形能都不应改变。

## 本题的目的

检验有限元方法在单元层面的计算是否与物理意义相符？

## 几个要点 —— 考虑3结点三角形单元

- 单元$e$的变形能为

$$
U_e=\dfrac{1}{2}\boldsymbol{a}^T\boldsymbol{K}_e\boldsymbol{a}
$$

- 应力、应变、位移关系

$$
\boldsymbol{\sigma}=\boldsymbol{D\varepsilon}=\boldsymbol{DBa}=\boldsymbol{Sa}
$$

- 应变矩阵
$$
\boldsymbol{B}
=
\dfrac{1}{2A}
\left[\begin{array}{cc|cc|cc}
b_i  & 0   & b_j  & 0   & b_m  & 0  \\
0   & c_i  & 0   & c_j  & 0   & c_m\\
c_i  & b_i  & c_j  & b_j  & c_m  & b_m
\end{array}\right]
$$

- 应力矩阵
$$
\boldsymbol{S}=\boldsymbol{DB}
$$

- 单元刚度矩阵
$$
\boldsymbol{K}_e=\boldsymbol{B}^T\boldsymbol{D}\boldsymbol{B}At
$$

## 单元发生移动$(x_0,y_0)$时

假设单元平移后的新应变矩阵变成
$$
\bar{\boldsymbol{B}}
=
\dfrac{1}{2\bar{A}}
\left[\begin{array}{cc|cc|cc}
\bar{b}_i  & 0   & \bar{b}_j  & 0   & \bar{b}_m  & 0  \\
0   & \bar{c}_i  & 0   & \bar{c}_j  & 0   & \bar{c}_m\\
\bar{c}_i  & \bar{b}_i  & \bar{c}_j  & \bar{b}_j  & \bar{c}_m  & \bar{b}_m
\end{array}\right]
$$
不难得到
$$
\bar{A}=
\left|\begin{array}{ccc}
	1 & x_i+x_0 & y_i+y_0\\
	1 & x_j+x_0 & y_j+y_0\\
	1 & x_m+x_0 & y_m+y_0
\end{array}\right|
=
\left|\begin{array}{ccc}
	1 & x_i & y_i\\
	1 & x_j & y_j\\
	1 & x_m & y_m
\end{array}\right|
=A
$$
以及
$$
\begin{array}{l}
	\bar{b}_i=-\left|\begin{array}{cc}
		1 & y_j+y_0\\
		1 & y_m+y_0
		\end{array}\right|
	   =y_j-y_m
	   =b_i\\
	\bar{b}_j=-\left|\begin{array}{cc}
		1 & y_m+y_0\\
		1 & y_i+y_0
		\end{array}\right|
	   =y_m-y_i
	   =b_j\\
	\bar{b}_m=-\left|\begin{array}{cc}
		1 & y_i+y_0\\
		1 & y_j+y_0
		\end{array}\right|
	   =y_i-y_j
	   =b_m
\end{array}
$$
和
$$
\begin{array}{l}
	\bar{c}_i=\left|
		\begin{array}{cc}
			1 & x_j+x_0\\
			1 & x_m+x_0
		\end{array}
		\right|
		=-x_j+x_m
      =c_i\\
	\bar{c}_j=\left|
		\begin{array}{cc}
			1 & x_m+x_0\\
			1 & x_i+x_0
		\end{array}
		\right|
		=-x_m+x_i
     =c_j\\
	\bar{c}_m=\left|
		\begin{array}{cc}
			1 & x_i+x_0\\
			1 & x_j+x_0
		\end{array}
		\right|
		=-x_i+x_j=c_m
\end{array}
$$

## 结论

- 应变矩阵不变
$$
\bar{\boldsymbol{B}}=\boldsymbol{B}
$$

- 应力矩阵不变
$$
\bar{\boldsymbol{S}}=\boldsymbol{D}\bar{\boldsymbol{B}}=\boldsymbol{DB}=\boldsymbol{S}
$$

- 单元刚度矩阵不变
$$
\bar{\boldsymbol{K}}_e=\bar{\boldsymbol{B}}^T\boldsymbol{D}\bar{\boldsymbol{B}}\bar{A}t=\boldsymbol{K}_e
$$

## 单元发生转动时

假定是绕原点$O$，（逆时针）转动角度为$\theta$，假设新的单元结点坐标为
- $(\tilde{x}_i,\tilde{y}_i)$
- $(\tilde{x}_j,\tilde{y}_j)$
- $(\tilde{x}_m,\tilde{y}_m)$

新老坐标关系为
$$
\left\{
\begin{array}{l}
\tilde{x}_\alpha\\
\tilde{y}_\alpha
\end{array}
\right\}
=
\left[\begin{array}{cc}
\cos\theta & -\sin\theta\\
\sin\theta & \cos\theta
\end{array}\right]
\left\{
\begin{array}{l}
{x}_\alpha\\
{y}_\alpha
\end{array}
\right\},\quad
\alpha = i,j,m
$$

假设单元旋转后的新应变矩阵变成
$$
\tilde{\boldsymbol{B}}
=
\dfrac{1}{2\tilde{A}}
\left[\begin{array}{cc|cc|cc}
\tilde{b}_i  & 0   & \tilde{b}_j  & 0   & \tilde{b}_m  & 0  \\
0   & \tilde{c}_i  & 0   & \tilde{c}_j  & 0   & \tilde{c}_m\\
\tilde{c}_i  & \tilde{b}_i  & \tilde{c}_j  & \tilde{b}_j  & \tilde{c}_m  & \tilde{b}_m
\end{array}\right]
$$
这时仍有
$$
\tilde{A}=
\left|\begin{array}{ccc}
	1 & \tilde{x_i} & \tilde{y_i}\\
	1 & \tilde{x_j} & \tilde{y_j}\\
	1 & \tilde{x_m} & \tilde{y_m}
\end{array}\right|
=
\left|\begin{array}{ccc}
	1 & x_i & y_i\\
	1 & x_j & y_j\\
	1 & x_m & y_m
\end{array}\right|
\left|\begin{array}{ccc}
	1 &   0      & 0      \\
	0 &  \cos\theta & \sin\theta   \\
	0 &  -\sin\theta & \cos\theta
\end{array}\right|
=A
$$
但一般情况下
$$
\begin{array}{l}
	\tilde{b}_i\neq b_i\\
	\tilde{b}_j\neq b_j\\
	\tilde{b}_m\neq b_m
\end{array}
$$
和
$$
\begin{array}{l}
	\tilde{c}_i\neq c_i\\
	\tilde{c}_j\neq c_j\\
	\tilde{c}_m\neq c_m
\end{array}
$$

## 说明 —— 一般情况下

- 新老应变矩阵不同
- 新老应力矩阵不同
- 新老单元刚度矩阵不同
- 新老单元结点荷载向量不同

## 特殊情况 —— 如果旋转180度

这时新老坐标关系为
$$
\left\{
\begin{array}{l}
\tilde{x}_i\\
\tilde{x}_j\\
\tilde{x}_m
\end{array}
\right\}
=
\left\{
\begin{array}{l}
-{x}_i\\
-{x}_j\\
-{x}_m
\end{array}
\right\}
,\quad\quad
\left\{
\begin{array}{l}
\tilde{y}_i\\
\tilde{y}_j\\
\tilde{y}_m
\end{array}
\right\}
=
\left\{
\begin{array}{l}
-{y}_i\\
-{y}_j\\
-{y}_m
\end{array}
\right\}
$$

于是
$$
\tilde{b}_\alpha=-b_\alpha, \quad \tilde{c}_\alpha=-c_\alpha, \quad \alpha = i,j,m
$$

- 应变矩阵
$$
\tilde{\boldsymbol{B}}
=
\dfrac{1}{2\tilde{A}}
\left[\begin{array}{cc|cc|cc}
\tilde{b}_i  & 0   & \tilde{b}_j  & 0   & \tilde{b}_m  & 0  \\
0   & \tilde{c}_i  & 0   & \tilde{c}_j  & 0   & \tilde{c}_m\\
\tilde{c}_i  & \tilde{b}_i  & \tilde{c}_j  & \tilde{b}_j  & \tilde{c}_m  & \tilde{b}_m
\end{array}\right]
=
-\boldsymbol{B}
$$

- 应力矩阵
$$
\tilde{\boldsymbol{S}}=\boldsymbol{D}\tilde{\boldsymbol{B}}=-\boldsymbol{DB}=-\boldsymbol{S}
$$

- 单元刚度矩阵
$$
\tilde{\boldsymbol{K}}_e
=\tilde{\boldsymbol{B}}^T\boldsymbol{D}\tilde{\boldsymbol{B}}\tilde{A}t
=(-\boldsymbol{B})^T\boldsymbol{D}(-\boldsymbol{B})At
=\boldsymbol{K}_e
$$

## 一般的情况

- 3结点三角形单元
- 4结点矩形单元
- 高次单元（三角形、四边形等）
- 单元的结点数和形状都不确定

理论上，都能得到相似的结论，不过，证明困难。

可以采用等参元的办法，对坐标进行插值。
